This project is a simple Python-based stock analysis tool built using:


*   yfinance for downloading real-time stock market data
*   pandas for data manipulation and moving average calculations
*   matplotlib for visualizing stock performance over the last 5 years

Features:


*   Input any stock ticker (e.g., AAPL, TSLA,MSFT)
*   Automatically fetches 5 years of daily stock data
*   Plots:
          *   Daily closing price
          *   50-day and 200-day moving averages
Uses:

Useful for entry-level financial analysis, understanding price trends and learning how to work with real financial data using Python.




















```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Ask user for stock ticker
ticker = input("Enter stock ticker (e.g., AAPL): ").upper()

# Download 5 years of historical data
data = yf.download(ticker, period="5y")

# Calculate moving averages
data["50_MA"] = data["Close"].rolling(window=50).mean()
data["200_MA"] = data["Close"].rolling(window=200).mean()

# Plot the price and moving averages
plt.figure(figsize=(14, 7))
plt.plot(data["Close"], label="Close Price", color='blue')
plt.plot(data["50_MA"], label="50-day MA", color='orange')
plt.plot(data["200_MA"], label="200-day MA", color='green')
plt.title(f"{ticker} Stock Price (Last 5 Years)")
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.legend()
plt.grid(True)
plt.show()
```
